```python
import pandas as pd
```

```python
csv = pd.read_csv("tcga-brca-survival__.csv")
csv
```

```python
csv.drop(labels=["Sample ID", "Image Path"], axis=1, inplace=True)
csv
```

```python
csv["Overall Survival Status"] = csv["Overall Survival Status"].apply(lambda x: "event" if x ==1 else "censored")
csv
```

```python
csv.rename({"Patient ID": "case_id", "Image ID": "slide_id", "Overall Survival Status": "label", "Overall Survival (Months)": "duration"},axis=1, inplace=True)
csv
```

```python
hospitals_to_exclude = ["TCGA-A1", "TCGA-B6", "TCGA-E9"]
csv = csv[~csv['case_id'].str.contains('|'.join(hospitals_to_exclude))]
csv
```

```python
csv.to_csv("tcga-brca-survival.csv", index=False)
```
